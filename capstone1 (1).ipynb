```python
#Calculate present change of Apple
 
 
#Function to calculate precent change
def percent_change(v1,v2):
    percent = ((v2-v1)/ abs(v1))*100
    #print (percent)
    return round(percent,3)
```

```python
#returns open/close
def addnum(v):
    return percent_change(v.open, v.close)

```

```python
apple_df = df1[df1.Name == 'AAPL']
```

```python
#shows the argument in the dataset
apple_df['percent'] = apple_df.apply(lambda x : percent_change(x.open, x.close), axis=1)
```

```python
apple_df.head()
```

```python
import statistics
apple_mean=statistics.mean(apple_df['volume'])
statement = "The avarage volume of Apple from 2013-2018 is:"
print (statement, int(apple_mean))

#Calculate total return 
start_price= apple_df.loc[AAPL.date=='2013-02-08']['open']
end_price= apple_df.loc[AAPL.date == '2018-02-07']['open']
total_return= (int(end_price)-int(start_price))/(int(start_price))
statement1= "The total return of the Apple stock from 2013-2018 is:"
print(statement1,round(total_return,2))

```

```python
#Graphing Data set
apple_df.date = pd.to_datetime(apple_df['date'], format='%Y-%m-%d' )
apple_date = apple_df.date.tolist()
apple_percent = apple_df.percent.cumsum().tolist()
plt.title('Apple Percent Change 2013-2018')
plt.xlabel('Years')
plt.ylabel('Percent Change')
plt.plot(apple_date,apple_percent)
```
